```python
import ollama
import ipywidgets as widgets
```

```python
text_area = widgets.Textarea(
    value='Hello World',
    placeholder='Type something',
    description='String:',
    disabled=False
)

button = widgets.Button(
    value=False,
    description='Click me',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)
```

```python
def continue_conversation(messages, temperature=0):
    response = ollama.chat(model='mistral',
        messages=messages,
       options={ "temperature":temperature}
    )
    #print(str(response.choices[0].message["content"]))
    return response['message']['content']
```

```python
#Creating the prompt
#read and understand it.
context = [ {'role':'system', 'content':"""
Act as an OrderBot, you work collecting orders in a delivery only fast food restaurant called
My Dear Frankfurt. \
First welcome the customer, in a very friendly way, then collects the order. \
You wait to collect the entire order, beverages included \
then summarize it and check for a final \
time if everything is ok or the customer wants to add anything else. \
Finally you collect the payment.\
Make sure to clarify all options, extras and sizes to uniquely \
identify the item from the menu.\
You respond in a short, very friendly style. \
The menu includes \
burger  12.95, 10.00, 7.00 \
frankfurt   10.95, 9.25, 6.50 \
sandwich   11.95, 9.75, 6.75 \
fries 4.50, 3.50 \
salad 7.25 \
Toppings: \
extra cheese 2.00, \
mushrooms 1.50 \
martra sausage 3.00 \
canadian bacon 3.50 \
romesco sauce 1.50 \
peppers 1.00 \
Drinks: \
coke 3.00, 2.00, 1.00 \
sprite 3.00, 2.00, 1.00 \
vichy catalan 5.00 \
"""} ]


def add_prompts_conversation(_):
    #Get the value introduced by the user
    if (text_area.value == ""):
        prompt = "Hi"
    else:
        prompt = text_area.value
    text_area.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    return f'User:  {prompt}\n Assistant:  {response}\n'


```

```python

panels = []
response_box = widgets.HBox([ widgets.Label("Welcome to Good Burger Home of the Good Burger Can I take your order?")])
def on_button_clicked(b):
    b = add_prompts_conversation(b)
    print(b)
    response_box.children += widgets.Box((b,))
button.on_click(on_button_clicked)
display(text_area)
display( button)
display(response_box)
```
